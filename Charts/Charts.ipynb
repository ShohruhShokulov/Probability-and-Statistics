```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('responses.csv')
favorite_genre = df['3.  What genre of books do you enjoy the most?'].value_counts()
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))


wedges, texts, autotexts = ax1.pie(favorite_genre, autopct='%1.1f%%', 
                                   startangle=90, labels=favorite_genre.index)
ax1.legend(wedges, favorite_genre.index, title='Genres', 
           loc='upper left', bbox_to_anchor=(1, 0, 0.5, 1))
ax1.set_title('Favorite Book Genre Distribution')

ax1.text(0.5, -0.1, f"Sample Size: {len(df)}", horizontalalignment='center', 
         verticalalignment='center', transform=ax1.transAxes)
colors = [w.get_facecolor() for w in wedges]

ax2.bar(favorite_genre.index, favorite_genre.values, color=colors)
ax2.set_title('Number of Responders by Genre')

for i, v in enumerate(favorite_genre.values):
    ax2.text(i, v + 0.1, str(v), ha='center', va='bottom')

plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('responses.csv')


preferred_medium = df['4.  What\'s your preferred reading medium( physical books, e-books, or audiobooks)?'].value_counts()


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6)) 


explode = (0.1, 0, 0) 
if len(explode) != len(preferred_medium):
    explode = tuple([0.1] + [0] * (len(preferred_medium) - 1))

pie_chart_colors = ['aqua', 'springgreen', 'salmon','wheat']
bar_chart_colors = ['aqua', 'springgreen', 'salmon','wheat']

wedges, texts, autotexts = ax1.pie(preferred_medium, autopct='%1.1f%%', startangle=90, labels=preferred_medium.index, explode=explode, counterclock=False, colors=pie_chart_colors)
ax1.legend(wedges, preferred_medium.index, title='Reading Mediums', loc='upper left', bbox_to_anchor=(1, 0, 0.5, 1))
ax1.set_title('Preferred Reading Medium Distribution')


ax1.text(0.5, -0.1, f"Sample Size: {len(df)}", horizontalalignment='center', verticalalignment='center', transform=ax1.transAxes)


colors = [w.get_facecolor() for w in wedges]


ax2.bar(preferred_medium.index, preferred_medium.values, color=bar_chart_colors)
ax2.set_title('Number of Responders by Reading Medium')


for i, v in enumerate(preferred_medium.values):
    ax2.text(i, v + 0.1, str(v), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
